```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset (clean data)
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
# define the data/predictors as the pre-set feature names
X = pd.DataFrame(data, columns=['production_budget_usd'])
# Put the target in another DataFrame
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $') 
plt.ylim(0, 3_000_000_000)
plt.xlim(0, 450_000_000)
plt.show()
```

```python
regression = LinearRegression()
# fit the features and labels (x for features and y for labels)
regression.fit(X, y) 
```

```python
# slope coefficient theta_1
regression.coef_
```

```python
regression.intercept_ #intercept
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color = 'red', linewidth = 4)
plt.title('Film vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $') 
plt.ylim(0, 3_000_000_000)
plt.xlim(0, 450_000_000)
plt.show()
```

```python
#measure goodness of fit
regression.score(X, y)
```
